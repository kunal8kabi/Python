Enter a number and check whether it's armstrong or not.

Armstrong number is a n digit number which is equal to sum of nth power of each digit.

For example 123 = 1^3 + 2^3 + 3^3

```
import numpy as np
num = int(input('Enter your number :'))
num1 = num
num2 = num
sum1 = 0
count = 0
if int(num2 / 10) != 0:
    while num2 != 0:  #to generate the power value of each digit
        num2 = int(num2 / 10)
        count = count + 1

    while num1 != 0:
        for j in range (1,count+1):
            i = num1 % 10
            sum1 = sum1 + np.power(i,count)
            num1 = int(num1 / 10)
    if sum1 == num:
        print(num,'is a armstrong number.')
    else:
        print(num,'is not a armstrong number.')

else:
    print(num, 'is a armstrong number.')
```
